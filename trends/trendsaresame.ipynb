```python
import xarray as xr
import numpy as np
```

```python
old_nc = xr.open_dataset("/Users/madisonrichardson/netpp/data/trends/noaa20/productivity_data_trend_month_noaa20_9km_2018_2022_050percent_NEWCDL.nc")
new_nc = xr.open_dataset("/Users/madisonrichardson/netpp/data/trends/productivity_data_trend_month_modis_9km_2018_2022_050percent_NEWCDL2.nc")
```

```python
print(f"Variables in old dataset: {list(old_nc.data_vars.keys())}")
print(f"Variables in new dataset: {list(new_nc.data_vars.keys())}")
```

```python
for var in old_nc.data_vars:
    print(f"🔍 Checking variable: {var}")
    print("Old Data Mean:", old_nc[var].mean().item())
    print("New Data Mean:", new_nc[var].mean().item())
    print("Difference:", (old_nc[var] - new_nc[var]).mean().item(), "\n")
```
