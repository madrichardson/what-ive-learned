```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the matrix text content
labels = [
    ["Interim: ns\nLegacy: n.s.", "Interim: ns\nLegacy: > 0", "Interim: ns\nLegacy: < 0"],
    ["Interim: > 0\nLegacy: n.s.", "Interim: > 0\nLegacy: > 0", "Interim: > 0\nLegacy: < 0"],
    ["Interim: < 0\nLegacy: n.s.", "Interim: < 0\nLegacy: > 0", "Interim: < 0\nLegacy: < 0"]
]
```

```python
# Matrix dimensions
nrows, ncols = 3, 3
```

```python
# Create a uniform matrix (values don't matter since no colormap will be used)
data = np.ones((nrows, ncols))

fig, ax = plt.subplots(figsize=(6, 6))

# Use a single facecolor (e.g., light gray)
table = ax.imshow(data, cmap="Greys", vmin=0, vmax=1)
table.set_visible(False)  # hide heatmap itself, we'll use Rectangle patches instead

# Add uniform rectangles with borders
for i in range(nrows):
    for j in range(ncols):
        rect = plt.Rectangle((j - 0.5, i - 0.5), 1, 1, fill=True, color="whitesmoke", edgecolor="black")
        ax.add_patch(rect)
        ax.text(j, i, labels[i][j], ha="center", va="center", fontsize=10)

# Set axis labels
ax.set_xticks(np.arange(ncols))
ax.set_yticks(np.arange(nrows))
ax.set_xticklabels(["Legacy: n.s.", "Legacy: > 0", "Legacy: < 0"], rotation=45, ha="right")
ax.set_yticklabels(["Interim: n.s.", "Interim: > 0", "Interim: < 0"])

# Formatting
ax.set_title("Contingency Table Pair Combinations", pad=20)
ax.tick_params(left=False, bottom=False)
ax.set_xticks(np.arange(-.5, ncols, 1), minor=True)
ax.set_yticks(np.arange(-.5, nrows, 1), minor=True)
ax.grid(False)
ax.set_xlim(-0.5, ncols - 0.5)
ax.set_ylim(nrows - 0.5, -0.5)

plt.tight_layout()
plt.show()
```

```python
# Cell labels
cell_text = [
    ["Interim: n.s.\nLegacy: n.s.", "Interim: n.s.\nLegacy: > 0", "Interim: n.s.\nLegacy: < 0"],
    ["Interim: > 0\nLegacy: n.s.", "Interim: > 0\nLegacy: > 0", "Interim: > 0\nLegacy: < 0"],
    ["Interim: < 0\nLegacy: n.s.", "Interim: < 0\nLegacy: > 0", "Interim: < 0\nLegacy: < 0"]
]

# Axis labels
x_labels = ["Legacy: n.s.", "Legacy: > 0", "Legacy: < 0"]
y_labels = ["Interim: n.s.", "Interim: > 0", "Interim: < 0"]

fig, ax = plt.subplots(figsize=(6, 6))

# Draw light blue boxes with white borders
for i in range(3):
    for j in range(3):
        rect = plt.Rectangle(
            (j, i), 1, 1,
            facecolor="#e9f1f8",       # light blue
            edgecolor="white",         # inner white border
            linewidth=1.5
        )
        ax.add_patch(rect)
        ax.text(j + 0.5, i + 0.5, cell_text[i][j], ha="center", va="center", fontsize=10)

# Draw outer black border (frame)
outer_rect = plt.Rectangle((0, 0), 3, 3, fill=False, edgecolor="black", linewidth=2)
ax.add_patch(outer_rect)

# Set limits and aspect ratio
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)
ax.set_aspect('equal')

# Set tick positions and labels
ax.set_xticks([0.5, 1.5, 2.5])
ax.set_yticks([0.5, 1.5, 2.5])
ax.set_xticklabels(x_labels, fontsize=11)
ax.set_yticklabels(y_labels, fontsize=11)

# Remove outer ticks and axis spines
ax.tick_params(top=False, bottom=False, left=False, right=False)
for spine in ax.spines.values():
    spine.set_visible(False)

# Invert y-axis so matrix reads top-down
ax.invert_yaxis()

# Optional title
ax.set_title("Contingency Table Pair Combinations", pad=20)

plt.tight_layout()
plt.savefig("/Users/madisonrichardson/netpp/work/Workflow/stats/contingency_matrix/Contingency_figure.png", dpi=300, bbox_inches="tight")
plt.show()
```
